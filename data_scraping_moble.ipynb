```python
model = []
price = []
ranting = []
ram = []
display = []
camera = []
battery = []
processor = []
warrenty = []
for i in range(1,14):

    url = f"https://www.flipkart.com/mobiles/pr?sid=tyy%2C4io&p%5B%5D=facets.brand%255B%255D%3DOPPO&param=1001&ctx=eyJjYXJkQ29udGV4dCI6eyJhdHRyaWJ1dGVzIjp7InRpdGxlIjp7Im11bHRpVmFsdWVkQXR0cmlidXRlIjp7ImtleSI6InRpdGxlIiwiaW5mZXJlbmNlVHlwZSI6IlRJVExFIiwidmFsdWVzIjpbIk9wcG8gc21hcnRwaG9uZXMiXSwidmFsdWVUeXBlIjoiTVVMVElfVkFMVUVEIn19fX19&wid=46.productCard.PMU_V2_41&page={i}"
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'lxml')
    phones = soup.find_all('div', class_='_2kHMtA')
    count = 0
    for phone in phones:
        try:
            model.append(phone.find('div',class_='_4rR01T').text)
        except:
            model.append(np.NAN)   
        try:     
            price.append(phone.find('div',class_='_30jeq3 _1_WHN1').text)
        except:
            price.append(np.NAN)
        try:    
            ranting.append(phone.find('div', class_='_3LWZlK').text)
        except:
            ranting.append(np.NAN)
        
        try:
            ram.append(phone.find('ul', class_='_1xgFaf').find_all('li')[0].text)
        except:
            ram.append(np.NAN)

        try:    
            display.append(phone.find('ul', class_='_1xgFaf').find_all('li')[1].text)
        except:
            display.append(np.NAN)

        try:        
            camera.append(phone.find('ul', class_='_1xgFaf').find_all('li')[2].text)
        except:
            camera.append(np.NAN)

        try:        
            processor.append(phone.find('ul', class_='_1xgFaf').find_all('li')[3].text)
        except:
            processor.append(np.NAN) 

        try:       
            battery.append(phone.find('ul', class_='_1xgFaf').find_all('li')[4].text)
        except:
            battery.append(np.NAN)

        try:    
            warrenty.append(phone.find('ul', class_='_1xgFaf').find_all('li')[5].text)
        except:
            warrenty.append(np.NAN)    
        
```

```python
oppo = pd.DataFrame({'model':model,
                       'Price':price,
                       'ranting':ranting,
                       'ram':ram,
                       'display':display,
                       'camera':camera,
                       'battery':processor,
                       'processor':battery,
                       'warrenty':warrenty})
```
